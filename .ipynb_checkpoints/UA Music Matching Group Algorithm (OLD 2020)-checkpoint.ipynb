```python
group_id = {}
i = -1
for group in groups:
    i += 1
    for pair in group:
        group_id[pair] = i
print(group_id)
```

```python
person_id = {i: [] for i in range(len(groups))}
total_id = 0
for index, row in df.iterrows():
    if (row['TopG'].strip(), row['SecondG'].strip()) in group_id:
        person_id[group_id[(row['TopG'].strip(), row['SecondG'].strip())]].append(row['Kerb'])
        total_id += 1
        
print(total_id)
print(person_id)
```

```python
import random

#now lets split the bigger genres randomly
big_groups = [('Rap', 'Pop'), ('Pop', 'Rock'), ('Pop', 'Modern Rock')]
biggest_group = [('Pop', 'Rap')]
j = len(groups)
for g in big_groups:
    full_group = []
    for index, row in df.iterrows():
        if (row['TopG'].strip(), row['SecondG'].strip()) == g:
            full_group.append(row['Kerb'])
            total_id += 1
    random.shuffle(full_group)
    person_id[j] = full_group[:int(len(full_group)/2)]
    person_id[j+1] = full_group[int(len(full_group)/2):]
    j += 2
```

```python
biggest_group = [('Pop', 'Rap')]
for g in biggest_group:
    full_group = []
    for index, row in df.iterrows():
        if (row['TopG'].strip(), row['SecondG'].strip()) == g:
            full_group.append(row['Kerb'])
            total_id += 1
    random.shuffle(full_group)
    person_id[j] = full_group[:int(len(full_group)/4)]
    person_id[j+1] = full_group[int(len(full_group)/4):2*int(len(full_group)/4)]
    person_id[j+2] = full_group[2*int(len(full_group)/4):3*int(len(full_group)/4)]
    person_id[j+3] = full_group[3*int(len(full_group)/4):]
    j += 4
    
print(person_id)
```

```python
from collections import defaultdict
artistDict = defaultdict(int)

for index, row in df.iterrows():
    if (row['TopG'].strip(), row['SecondG'].strip()) == ('Pop', 'Rap') or (row['TopG'].strip(), row['SecondG'].strip()) == ('Rap', 'Pop'):
        artistDict[row['FirstA']] += 1
        artistDict[row['SecondA']] += 1
        artistDict[row['ThirdA']] += 1
        #artistDict[row['FourthA']] += 1
        #artistDict[row['FifthA']] += 1
        
#print(artistDict)

for w in sorted(artistDict, key=artistDict.get, reverse=True):
    print(w, artistDict[w])
```
